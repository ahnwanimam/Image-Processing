```python
from PyQt5 import QtCore, QtGui, QtWidgets
from PyQt5.QtWidgets import QMainWindow, QApplication, QMessageBox, QFileDialog
import cv2
import numpy as np


class Ui_MainWindow(object):
    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.resize(436, 600)
        
        self.centralwidget = QtWidgets.QWidget(MainWindow)
        self.centralwidget.setObjectName("centralwidget")
        
        self.deleteWin = QtWidgets.QPushButton(self.centralwidget)
        self.deleteWin.setGeometry(QtCore.QRect(170, 500, 75, 23))
        self.deleteWin.setObjectName("deleteWin")
        
        self.create = QtWidgets.QPushButton(self.centralwidget)
        self.create.setGeometry(QtCore.QRect(80, 440, 75, 23))
        self.create.setObjectName("create")
        
        self.load = QtWidgets.QPushButton(self.centralwidget)
        self.load.setGeometry(QtCore.QRect(270, 440, 75, 23))
        self.load.setObjectName("load")
        
        self.Draw = QtWidgets.QLabel(self.centralwidget)
        self.Draw.setGeometry(QtCore.QRect(150, 0, 181, 61))
        font = QtGui.QFont()
        font.setFamily("KoPub돋움체 Bold")
        font.setPointSize(20)
        self.Draw.setFont(font)
        self.Draw.setObjectName("Draw")
        
        self.shape = QtWidgets.QLabel(self.centralwidget)
        self.shape.setGeometry(QtCore.QRect(300, 50, 91, 21))
        font = QtGui.QFont()
        font.setFamily("KoPub돋움체 Bold")
        font.setPointSize(11)
        self.shape.setFont(font)
        self.shape.setObjectName("shape")
        
        self.color = QtWidgets.QLabel(self.centralwidget)
        self.color.setGeometry(QtCore.QRect(70, 50, 51, 21))
        font = QtGui.QFont()
        font.setFamily("KoPub돋움체 Bold")
        font.setPointSize(11)
        self.color.setFont(font)
        self.color.setObjectName("color")
        
        self.red = QtWidgets.QToolButton(self.centralwidget)
        self.red.setGeometry(QtCore.QRect(10, 80, 41, 41))
        self.red.setText("")
        icon = QtGui.QIcon()
        icon.addPixmap(QtGui.QPixmap("images/red.png"), QtGui.QIcon.Normal, QtGui.QIcon.Off)
        self.red.setIcon(icon)
        self.red.setIconSize(QtCore.QSize(50, 50))
        self.red.setCheckable(True)
        self.red.setObjectName("red")
        
        self.green = QtWidgets.QToolButton(self.centralwidget)
        self.green.setGeometry(QtCore.QRect(60, 80, 41, 41))
        self.green.setText("")
        icon1 = QtGui.QIcon()
        icon1.addPixmap(QtGui.QPixmap("images/green.png"), QtGui.QIcon.Normal, QtGui.QIcon.Off)
        self.green.setIcon(icon1)
        self.green.setIconSize(QtCore.QSize(27, 27))
        self.green.setCheckable(True)
        self.green.setObjectName("green")
        
        self.blue = QtWidgets.QToolButton(self.centralwidget)
        self.blue.setGeometry(QtCore.QRect(110, 80, 41, 41))
        self.blue.setText("")
        icon2 = QtGui.QIcon()
        icon2.addPixmap(QtGui.QPixmap("images/blue.png"), QtGui.QIcon.Normal, QtGui.QIcon.Off)
        self.blue.setIcon(icon2)
        self.blue.setIconSize(QtCore.QSize(30, 30))
        self.blue.setCheckable(True)
        self.blue.setObjectName("blue")
        
        self.black = QtWidgets.QToolButton(self.centralwidget)
        self.black.setGeometry(QtCore.QRect(160, 80, 41, 41))
        self.red.setText("")
        icon = QtGui.QIcon()
        icon.addPixmap(QtGui.QPixmap("images/black.png"), QtGui.QIcon.Normal, QtGui.QIcon.Off)
        self.black.setIcon(icon)
        self.black.setIconSize(QtCore.QSize(50, 50))
        self.black.setCheckable(True)
        self.black.setObjectName("black")
        
        self.pen = QtWidgets.QToolButton(self.centralwidget)
        self.pen.setGeometry(QtCore.QRect(230, 80, 41, 41))
        self.pen.setIconSize(QtCore.QSize(50, 50))
        self.pen.setCheckable(True)
        self.pen.setObjectName("pen")
        
        self.line = QtWidgets.QToolButton(self.centralwidget)
        self.line.setGeometry(QtCore.QRect(280, 80, 41, 41))
        self.line.setIconSize(QtCore.QSize(50, 50))
        self.line.setCheckable(True)
        self.line.setObjectName("line")
        
        self.rect = QtWidgets.QToolButton(self.centralwidget)
        self.rect.setGeometry(QtCore.QRect(330, 80, 41, 41))
        self.rect.setIconSize(QtCore.QSize(50, 50))
        self.rect.setCheckable(True)
        self.rect.setObjectName("rect")
        
        self.circle = QtWidgets.QToolButton(self.centralwidget)
        self.circle.setGeometry(QtCore.QRect(380, 80, 41, 41))
        self.circle.setIconSize(QtCore.QSize(50, 50))
        self.circle.setCheckable(True)
        self.circle.setObjectName("circle")
        
        self.remove = QtWidgets.QToolButton(self.centralwidget)
        self.remove.setGeometry(QtCore.QRect(380, 130, 41, 41))
        self.remove.setIconSize(QtCore.QSize(50, 50))
        self.remove.setCheckable(True)
        self.remove.setObjectName("remove")
        
        self.photo = QtWidgets.QLabel(self.centralwidget)
        self.photo.setGeometry(QtCore.QRect(150, 160, 181, 61))
        font = QtGui.QFont()
        font.setFamily("KoPub돋움체 Bold")
        font.setPointSize(20)
        self.photo.setFont(font)
        self.photo.setObjectName("photo")
        
        self.blur = QtWidgets.QToolButton(self.centralwidget)
        self.blur.setGeometry(QtCore.QRect(190, 400, 41, 41))
        self.blur.setIconSize(QtCore.QSize(50, 50))
        self.blur.setCheckable(True)
        self.blur.setObjectName("blur")
        
        self.blur_hard = QtWidgets.QLabel(self.centralwidget)
        self.blur_hard.setGeometry(QtCore.QRect(180, 220, 91, 21))
        font = QtGui.QFont()
        font.setFamily("KoPub돋움체 Bold")
        font.setPointSize(11)
        self.blur_hard.setFont(font)
        self.blur_hard.setObjectName("blur_hard")
        
        self.blur_thick = QtWidgets.QLabel(self.centralwidget)
        self.blur_thick.setGeometry(QtCore.QRect(180, 300, 91, 21))
        font = QtGui.QFont()
        font.setFamily("KoPub돋움체 Bold")
        font.setPointSize(11)
        self.blur_thick.setFont(font)
        self.blur_thick.setObjectName("blur_thick")
        
        self.big = QtWidgets.QToolButton(self.centralwidget)
        self.big.setGeometry(QtCore.QRect(140, 330, 41, 41))
        self.big.setIconSize(QtCore.QSize(50, 50))
        self.big.setCheckable(True)
        self.big.setObjectName("big")
        
        self.nor = QtWidgets.QToolButton(self.centralwidget)
        self.nor.setGeometry(QtCore.QRect(190, 330, 41, 41))
        self.nor.setIconSize(QtCore.QSize(50, 50))
        self.nor.setCheckable(True)
        self.nor.setObjectName("nor")
        
        self.small = QtWidgets.QToolButton(self.centralwidget)
        self.small.setGeometry(QtCore.QRect(240, 330, 41, 41))
        self.small.setIconSize(QtCore.QSize(50, 50))
        self.small.setCheckable(True)
        self.small.setObjectName("small")
        
        self.hard = QtWidgets.QToolButton(self.centralwidget)
        self.hard.setGeometry(QtCore.QRect(140, 250, 41, 41))
        self.hard.setIconSize(QtCore.QSize(50, 50))
        self.hard.setCheckable(True)
        self.hard.setObjectName("hard")
        
        self.mid = QtWidgets.QToolButton(self.centralwidget)
        self.mid.setGeometry(QtCore.QRect(190, 250, 41, 41))
        self.mid.setIconSize(QtCore.QSize(50, 50))
        self.mid.setCheckable(True)
        self.mid.setObjectName("mid")
        
        self.soft = QtWidgets.QToolButton(self.centralwidget)
        self.soft.setGeometry(QtCore.QRect(240, 250, 41, 41))
        self.soft.setIconSize(QtCore.QSize(50, 50))
        self.soft.setCheckable(True)
        self.soft.setObjectName("soft")
        
        MainWindow.setCentralWidget(self.centralwidget)
        self.menubar = QtWidgets.QMenuBar(MainWindow)
        self.menubar.setGeometry(QtCore.QRect(0, 0, 436, 21))
        self.menubar.setObjectName("menubar")
        MainWindow.setMenuBar(self.menubar)
        self.statusbar = QtWidgets.QStatusBar(MainWindow)
        self.statusbar.setObjectName("statusbar")
        MainWindow.setStatusBar(self.statusbar)

        self.retranslateUi(MainWindow)
        self.create.clicked.connect(self.btn_create)
        self.deleteWin.clicked.connect(self.btn_deleteWin)
        self.pen.clicked.connect(self.btn_pen)
        self.line.clicked.connect(self.btn_line)
        self.rect.clicked.connect(self.btn_rect)
        self.circle.clicked.connect(self.btn_circle)
        self.remove.clicked.connect(self.btn_remove)
        self.black.clicked.connect(self.btn_black)
        self.red.clicked.connect(self.btn_red)
        self.green.clicked.connect(self.btn_green)
        self.blue.clicked.connect(self.btn_blue)
        self.load.clicked.connect(self.btn_load)
        self.blur.clicked.connect(self.btn_blur)
        self.hard.clicked.connect(self.btn_hard)
        self.mid.clicked.connect(self.btn_mid)
        self.soft.clicked.connect(self.btn_soft)
        self.big.clicked.connect(self.btn_big)
        self.nor.clicked.connect(self.btn_nor)
        self.small.clicked.connect(self.btn_small)
        QtCore.QMetaObject.connectSlotsByName(MainWindow)
        
    #그림판 생성 버튼을 눌렀을 떼
    def btn_create(self):
        global title, image   #title과 image 전역변수 참조
        image = np.full((300, 500, 3), (255, 255, 255), np.uint8)   # 삭제 후 다시 생성 했을 때 다시 그리는 그림판을 흰색으로 초기화
        cv2.imshow(title, image)    #흰 그림판을 화면에 나타낸다.
    
    #그림판 삭제 버튼을 눌렀을 때
    def btn_deleteWin(self):
        cv2.destroyAllWindows()     #생성한 그림판을 없앤다.
    
    #"검" 버튼을 눌렀을 때
    def btn_black(self):
        global color     #color 전역변수 참조
        self.black.setChecked(True)       #한번에 여러개의 색상 버튼이 클릭 되면 안되기 때문에 "검" 버튼 외에 
        self.red.setChecked(False)        #다른 버튼들을 False로 하고, "검" 버튼만 True로 설정.
        self.green.setChecked(False)
        self.blue.setChecked(False)
        
        color = (0, 0, 0)    #색을 검정색으로 만든다.
    
    #"빨" 버튼을 눌렀을 때
    def btn_red(self):
        global color     #color 전역변수 참조
        self.red.setChecked(True)         #마찬가지로 "빨" 버튼 외에 다른 버튼을 False로 설정.
        self.black.setChecked(False)
        self.green.setChecked(False)
        self.blue.setChecked(False)
        
        color = (0, 0, 255)   #색을 빨간색으로 만든다.
    
    #"초" 버튼을 눌렀을 때
    def btn_green(self):
        global color     #color 전역변수 참조
        self.green.setChecked(True)       # "초" 버튼 외에 다른 버튼을 False로 설정.
        self.red.setChecked(False)
        self.black.setChecked(False)
        self.blue.setChecked(False)
        
        color = (0, 255, 0)   #색을 초록색으로 만든다.
    
    #"파" 버튼을 눌렀을 때
    def btn_blue(self):
        global color    #color 전역변수 참조
        self.blue.setChecked(True)       # "파" 버튼 외에 다른 버튼을 False로 설정.
        self.red.setChecked(False)
        self.black.setChecked(False)
        self.green.setChecked(False)
        
        color = (255, 0, 0)  #색을 파란색으로 만든다.
    
    #"펜" 버튼을 눌렀을 때
    def btn_pen(self):
        self.pen.setChecked(True)      #펜 버튼 외에 다른 버튼을 False로 설정.
        self.line.setChecked(False)
        self.rect.setChecked(False)
        self.circle.setChecked(False)
        self.remove.setChecked(False)
        self.blur.setChecked(False)
        
        def onMouse(event, x, y, flags, param):
            global title, image, pt, color     #title, image, pt, color 전역변수 참조               

            if event == cv2.EVENT_MOUSEMOVE:
                if flags & cv2.EVENT_FLAG_LBUTTON:
                    if pt[0] < 0:
                        pt = (x, y)                             #마우스 왼쪽을 누르고 있는 동안 그린다.
                    else:  
                        cv2.line(image, pt, pt, color, 2)   #cv2.line 의 시작과 끝 좌표를 같게 해 점을 찍는다. 마우르를 
                        cv2.imshow(title, image)            # 누르는 동안 계속 점을 찍어 자유롭게 그리도록 하였다.
                        pt = (-1, -1)  
             
        cv2.setMouseCallback(title, onMouse)            
    
    #"직선" 버튼을 눌렀을 때
    def btn_line(self):
        self.line.setChecked(True)      # 직선 버튼 외 모두 False 설정.
        self.rect.setChecked(False) 
        self.circle.setChecked(False)
        self.pen.setChecked(False)
        self.remove.setChecked(False)
        self.blur.setChecked(False)
        
        def onMouse(event, x, y, flags, param):   #교재 event_draw 참조. 과제로 따로 했던 드래그 방식이 아닌, 교재 방식 사용
            global title, image, pt, color     #title, image, pt, color 전역변수 참조               
            if event == cv2.EVENT_LBUTTONDOWN:   #마우스 왼쪽 버튼을 눌러 이벤트 처리
                if pt[0] < 0:           #첫 클릭이 없었다면
                    pt = (x, y)         #시작 좌표를 정해준다.
                else:                   #두번 째 클릭
                    cv2.line(image, pt, (x, y), color, 2)    #색상 버튼의 색상과 처음 클릭으로 얻은 시작좌표와 두번 째 좌표로 직선을 그린다 
                    cv2.imshow(title, image)      #화면에 나타낸다.
                    pt = (-1, -1)                 #좌표 초기화
               
        cv2.setMouseCallback(title, onMouse)         
    
    #"사각형" 버튼을 눌렀을 때
    def btn_rect(self):
        self.rect.setChecked(True)           # 사각형 버튼 외 모두 False 설정. 
        self.line.setChecked(False)
        self.circle.setChecked(False)
        self.pen.setChecked(False)
        self.remove.setChecked(False)
        self.blur.setChecked(False)
        
        def onMouse(event, x, y, flags, param):
            global title, image, pt, color        #title, image, pt, color 전역변수 참조              

            if event == cv2.EVENT_LBUTTONDOWN:  #마우스 왼쪽 버튼을 눌러 이벤트 처리
                if pt[0] < 0:        #아직 첫 클릭이 없으면
                    pt = (x, y)      #시작 좌표 설정
                else:     #그게 아니면(두번 째 클릭이라면)
                    cv2.rectangle(image, pt, (x, y), color, 2)  #색상버튼의 색상과 두 좌표로 사각형을 그린다.
                    cv2.imshow(title, image)     #화면에 나타낸다.
                    pt = (-1, -1)                #시작 좌표 초기화
               
        cv2.setMouseCallback(title, onMouse)
 
    #"원" 버튼을 눌렀을 때
    def btn_circle(self):
        self.circle.setChecked(True)       #원 버튼 외 모두 False 설정.
        self.line.setChecked(False)
        self.rect.setChecked(False)
        self.pen.setChecked(False)
        self.remove.setChecked(False)
        self.blur.setChecked(False)
        
        def onMouse(event, x, y, flags, param):
            global title, image, pt, color        #title, image, pt, color 전역변수 참조              

            if event == cv2.EVENT_LBUTTONDOWN:  #마우스 왼쪽 버튼을 눌러 이벤트 처리
                if pt[0] < 0:       #아직 첫 클릭이 없으면
                    pt = (x, y)     #시작 좌표로 설정
                else:
                    dx, dy = pt[0] - x, pt[1] - y       # 찻 시작좌표와 두번 째 좌표 사이의 거리
                    radius = int(np.sqrt(dx * dx + dy * dy))  #두 좌표 사이의 거리로 반지름을 구한다.
                    cv2.circle(image, pt, radius, color, 2)    #색상과 구한 반지름으로 원을 그린다.
                    cv2.imshow(title, image)    #화면에 나타낸다.
                    pt = (-1, -1)                       # 시작 좌표 초기화
        
        cv2.setMouseCallback(title, onMouse)
    
    #"지우개" 버튼을 눌렀을 때
    def btn_remove(self):
        self.remove.setChecked(True)
        self.line.setChecked(False)
        self.rect.setChecked(False)
        self.circle.setChecked(False)
        self.pen.setChecked(False)
        self.blur.setChecked(False)
        
        def onMouse(event, x, y, flags, param):
            global title, image, pt, color     #title, image, pt, color 전역변수 참조               

            if event == cv2.EVENT_MOUSEMOVE:
                if flags & cv2.EVENT_FLAG_LBUTTON:
                    if pt[0] < 0:
                        pt = (x, y)
                    else:  
                        cv2.line(image, pt, (x, y), (255,255,255), 10)   #그리기를 흰색으로 그려 지운다.
                        cv2.imshow(title, image)
                        pt = (-1, -1)  
             
        cv2.setMouseCallback(title, onMouse) 
        
    #"불러오기" 버튼을 눌렀을 때
    def btn_load(self):
        global title, image
        global filename
        filename = QFileDialog.getOpenFileNames(filter="")[0]
        
        if filename[0]:
            image = cv2.imread(filename[0], cv2.IMREAD_COLOR)
            image = cv2.resize(image, (500, 300))    #불러 오는 이미지 크기를 통일 시켰다.
            cv2.imshow(title, image)
    
    #"블러" 버튼을 눌렀을 때
    def btn_blur(self):
        self.green.setChecked(False)
        self.red.setChecked(False)
        self.black.setChecked(False)
        self.blue.setChecked(False)
        self.remove.setChecked(False)
        self.line.setChecked(False)
        self.rect.setChecked(False)
        self.circle.setChecked(False)
        self.pen.setChecked(False)
        global title, image, k, t
        def onMouse(event, x, y, flags, param):            
            if event == cv2.EVENT_LBUTTONDOWN:  #마우스 왼쪽 버튼을 눌러 이벤트 처리
                roi = image[y-t:y+t, x-t:x+t]      #양 좌표를 모두 -t, +t 만큼 영역을 지정해 준 후, 블러처리
                roi = cv2.blur(roi, (k,k))         # k는 블러의 세기, t는 블러의 굵기 모두 전역변수를 참조.
                image[y-t:y+t, x-t:x+t] = roi
                cv2.imshow(title, image)
        
        cv2.setMouseCallback(title, onMouse) 
   


    #<-------------------------------------버튼으로 각각 굵기와 세기를 지정해준다. ------------------------------------------------->
    def btn_hard(self):
        global k
        self.hard.setChecked(True)
        self.mid.setChecked(False)
        self.soft.setChecked(False)
        
        k = 15
        
    def btn_mid(self):
        global k
        self.mid.setChecked(True)
        self.hard.setChecked(False)
        self.soft.setChecked(False)
        
        k = 10
        
    def btn_soft(self):
        global k
        self.soft.setChecked(True)
        self.mid.setChecked(False)
        self.hard.setChecked(False)
        
        k = 5
        
    def btn_big(self):
        global t
        self.big.setChecked(True)
        self.nor.setChecked(False)
        self.small.setChecked(False)
        
        t = 15
        
    def btn_nor(self):
        global t
        self.nor.setChecked(True)
        self.big.setChecked(False)
        self.small.setChecked(False)
        
        t = 10
        
    def btn_small(self):
        global t
        self.small.setChecked(True)
        self.nor.setChecked(False)
        self.big.setChecked(False)
        
        t = 5
    #<-------------------------------------------------------------------------------------------------------------------------------->    
        

    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("MainWindow", "MainWindow"))
        self.deleteWin.setText(_translate("MainWindow", "삭제"))
        self.create.setText(_translate("MainWindow", "그림판 생성"))
        self.load.setText(_translate("MainWindow", "불러오기"))
        self.Draw.setText(_translate("MainWindow", "그림판 도구"))
        self.shape.setText(_translate("MainWindow", "   모양"))
        self.color.setText(_translate("MainWindow", " 색"))
        self.pen.setText(_translate("MainWindow", "펜"))
        self.line.setText(_translate("MainWindow", "직선"))
        self.rect.setText(_translate("MainWindow", "사각형"))
        self.circle.setText(_translate("MainWindow", "원"))
        self.remove.setText(_translate("MainWindow", "지우개"))
        self.photo.setText(_translate("MainWindow", " 보정 도구"))
        self.blur.setText(_translate("MainWindow", "블러"))
        self.blur_hard.setText(_translate("MainWindow", "블러 세기"))
        self.hard.setText(_translate("MainWindow", "강"))
        self.mid.setText(_translate("MainWindow", "중"))
        self.soft.setText(_translate("MainWindow", "약"))
        self.blur_thick.setText(_translate("MainWindow", "블러 굵기"))
        self.big.setText(_translate("MainWindow", "대"))
        self.nor.setText(_translate("MainWindow", "중"))
        self.small.setText(_translate("MainWindow", "소"))


if __name__ == "__main__":
    import sys
    global title, image, pt, color        #title, image, pt, color를 전역변수로 선언
    
    pt = (-1, -1)   #가장 처음 시작좌표 지정
    title = "Board"   #그림판 이름을 정해준다.
    color = (0, 0, 0)  #기본색상을 검정색으로 설정
    image = np.full((300, 500, 3), (255, 255, 255), np.uint8)
    
    app = QtWidgets.QApplication(sys.argv)
    MainWindow = QtWidgets.QMainWindow()
    ui = Ui_MainWindow()
    ui.setupUi(MainWindow)
    MainWindow.show()
    sys.exit(app.exec_())
```
